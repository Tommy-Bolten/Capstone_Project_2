```python
#finding unique values in each column to implement one hot encoding
for column in df1:
    print(f'{column} :{df1[column].unique()}')
    
```

```python
#Selecting columns where the data type is an object
def print_unique_col_values(df1):
    for column in df1:
        if df1[column].dtypes=='object':
            print(f'{column} :{df1[column].unique()}')
```

```python
print_unique_col_values(df1)
```

```python
#label encoding to zeros and ones
df1['gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
df1['gender'].dtypes
```

```python
#Replacing No internet service and No phone Service with No
df1 = df1.replace({'No internet service': 'No'})

df1.replace('No phone service', 'No',inplace=True)
```

```python
print_unique_col_values(df1)
```

```python
#Replacing Yes and no with ones and Zeros repectively
yes_no_columns = ['Partner', 'Dependents', 'PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                  'DeviceProtection', 'TechSupport','StreamingTV', 'StreamingMovies','PaperlessBilling','Churn']

for col in yes_no_columns:
    df1[col].replace({'Yes':1, 'No': 0}, inplace=True)
```

```python
print_unique_col_values(df1)
```

```python
#One hot encoding the 3 remaining columns
df2 = pd.get_dummies(data=df1, columns=['InternetService','Contract','PaymentMethod'], dtype =int)
df2.dtypes
```

```python
df2.head()
```

```python
categorical_columns = df2[['gender','Partner', 'Dependents', 'PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                  'DeviceProtection', 'TechSupport','StreamingTV', 'StreamingMovies','PaperlessBilling','Churn', 'InternetService_DSL','InternetService_Fiber optic',
'InternetService_No','Contract_Month-to-month','Contract_One year','Contract_Two year','PaymentMethod_Bank transfer (automatic)','PaymentMethod_Credit card (automatic)','PaymentMethod_Electronic check','PaymentMethod_Mailed check']]


df3 = pd.DataFrame(categorical_columns)
```

```python
# Verify that all columns in categorical_columns are present in the DataFrame
missing_columns = [col for col in categorical_columns if col not in df3.columns]
if missing_columns:
    print(f"Missing columns in DataFrame: {missing_columns}")
else:
    print("All columns are present in the DataFrame.")

# Loop through pairs of columns to generate crosstabs
if not missing_columns:
    for col1 in categorical_columns:
        for col2 in categorical_columns:
            if col1 != col2:
                crosstab = pd.crosstab(df3[col1], df3[col2])
                print(f'Crosstab between {col1} and {col2}:')
                print(crosstab)
                print('\n')
```

```python
#Performing Chi Squared test on categorical variables
contingency_table = pd.crosstab(df2['gender'],df2['Partner'])
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f'Chi-Square statistic: {chi2}')
print(f'p-value: {p}')
print(f'Degrees of freedom: {dof}')
print(f'Expected frequencies: \n{expected}')
```
